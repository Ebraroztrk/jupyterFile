```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function
def quadratic_function(x):
    return x**2

N = 100
x_values = np.random.uniform(low=-10, high=10, size=N)
y_values_actual = quadratic_function(x_values)


linear_coefficients = np.polyfit(x_values, y_values_actual, 1)
m, b = linear_coefficients

y_values_linear_approx = m * x_values + b

# Plot the quadratic function, noisy data, and linear approximation

plt.xlim([-10, 10])
plt.ylim([-5,100])
plt.scatter(x_values, y_values_actual)
plt.plot(x_values, y_values_linear_approx, label=f'Linear Approximation: y = {m:.2f}x + {b:.2f}', linestyle='-', color='red')
plt.title('Linear Approximation to Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

def generate_data(N):
    x = np.random.uniform(low=-10, high=10, size=N)
    y = x**2
    return x.reshape(-1, 1), y  # Reshape x to make it 2D

def perform_linear_regression(x, y):
    model = LinearRegression()
    model.fit(x, y)
    m = model.coef_[0]
    b = model.intercept_
    return m, b

def plot_results(x, y, m, b, actual_m, actual_b, title):
    plt.scatter(x, y, label='Actual Data')
    plt.plot(x, m*x + b, color='red', label=f'Linear Approximation: y = {m:.2f}x + {b:.2f}')
    plt.plot(x, actual_m*x + actual_b, color='green', label=f'analytical Approximation: y = {actual_m:.2f}x + {actual_b:.2f}')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

def quadratic_function(x):
    return x**2

def analytical_solution(N):
    x_values = np.random.uniform(low=-10, high=10, size=N)
    y_values_actual = quadratic_function(x_values)

    linear_coefficients = np.polyfit(x_values, y_values_actual, 1)
    m, b = linear_coefficients
    return m, b

def main():
    for k in range(1, 8):  # Vary k from 1 to 7
        N = 10**k
        x, y = generate_data(N)
        m, b = perform_linear_regression(x, y)
        actual_m, actual_b = analytical_solution(N)

        threshold = 1
        if abs(m - actual_m) < threshold and abs(b - actual_b) < threshold:
            print(f"For threshold = {threshold}, N = {N} and k = {k}, the linear function is sufficiently close.")
            plot_results(x, y, m, b, actual_m, actual_b, f'Linear Approximation for N = {N}')
            break

if __name__ == "__main__":
    main()
```
